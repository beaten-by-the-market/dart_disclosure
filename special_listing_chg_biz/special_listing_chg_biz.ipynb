# 기술성장기업 사업변경(주된 사업 변경) 분석

## 1. 기술성장기업 목록 불러오기

GitHub 레포 [`beaten-by-the-market/krx-data-api`](https://github.com/beaten-by-the-market/krx-data-api)의
`krx_data_api`를 설치해 KRX 정보데이터시스템의
**공모가 대비 주가수익률**(화면 20043) 데이터를 불러온다.

- `fetch("ipo_price_return")` : 코스닥 **기술성장기업**(`tecComType="TEC"`, 기본값)
- `fetch("ipo_price_return", tecComType="")` : 전체(기술성장기업 + 일반기업)

```python
# krx_data_api 패키지를 GitHub(beaten-by-the-market/krx-data-api)에서 설치
%pip install -q git+https://github.com/beaten-by-the-market/krx-data-api.git
```

```python
import pandas as pd
from krx_data_api import fetch

# 코스닥 기술성장기업 (tecComType="TEC" 기본값)
tech_df = fetch("ipo_price_return")

# 상장일(LIST_DD: "YYYY/MM/DD" 문자열) 기준 2023-01-01 ~ 2024-12-31 필터
list_dd = pd.to_datetime(tech_df["LIST_DD"], format="%Y/%m/%d")
mask = (list_dd >= "2023-01-01") & (list_dd <= "2024-12-31")
tech_df = tech_df[mask].copy()

print(f"2023.1.1~2024.12.31 상장 기술성장기업 {len(tech_df)}건")
tech_df.head()
```
